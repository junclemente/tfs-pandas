```python
import pandas as pd
```

## Select One Column from a DataFrame

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# Using dot notation. returns a series
nba.Team

# Output = None # Use this to prevent output
```

```python
# Using bracket notation. Will work with Columns that have spaces
nba["Name"]
```

## Select Two or More Columns from a DataFrame

```python
# Use a list inside the bracket notation. This will return a DataFrame. 
nba[["Name", "Team", "Height"]]
```

```python
# Columns can be rearranged: 
nba[["Name", "Height", "Team"]]
```

## Adding New Columns to an existing DataFrame

### Using braket notation

```python
nba["Sport"] = "Basketball"
nba.head()
```

```python
nba["League"] = "National Basketball Association"
nba
```

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

### Using .insert() method

```python
# .insert() will permanently modify the dataframe
nba.insert(3, column = "Sport", value = "Basketball")
nba
```
