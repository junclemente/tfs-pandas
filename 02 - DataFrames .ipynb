## Dropping Rows with Null Values

## Fill in Null Values with the .fillna() Method

```python
# Reset dataset
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# Fills null values with 0. 
nba.fillna(0)
```

```python
# Use the fillna method on a specific series to replace with more relevant information
# This example permanently  replaces all null values with "0". 
nba["Salary"].fillna(0, inplace = True)
nba["Salary"]
```

```python
# Use .fillna("No College") to replace all null values 
nba["College"].fillna("No College", inplace = True)
nba["College"]
```

## Converting datatypes of a series using .astype() Method

```python
# Using the .astype() method requires that there are no null values in the series 
nba.tail(3)
```

```python
#Reset and clean dataframe
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
nba.head(6)
```

```python
# Find information on dataypes
nba.dtypes
nba.info()
```

```python
nba["Salary"] = nba["Salary"].astype("int")
nba.head()
```

```python
nba.info()
```

```python
nba["Age"] = nba["Age"].astype("int")
nba["Number"] = nba["Number"].astype("int")
nba.info()
```

### "category" datatype

```python
# In the Position column, there are only 5 unique values. This is a perfect candidate for "category" type
# Converting datatypes to more appropriate type can reduce the size of memory usage
nba["Position"].nunique()
```

```python
nba["Position"] = nba["Position"].astype("category")
nba.info()
```

```python
nba["Team"].nunique()
```

```python
nba["Team"] = nba["Team"].astype("category")
nba.info()
```

## .sort_values Method

```python
# Reset dataset
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# By default, null values are placed at the end: parameter na_position = "last"
nba.sort_values("Salary")
```

```python
# Change location of null values using na_position parameter
nba.sort_values("Salary", na_position = "first")
```

```python
# Sorting by multiple columns by providing a list
nba.sort_values(["Team", "Name"])
```

```python
# Sorting and changing ascending to True/False by providing a list corresponding to the columns
nba.sort_values(["Team", "Name"], ascending = [True, False])
```

## Sort DataFrame with the .sort_index() Method

```python
nba.head(3)
```

```python
nba.sort_values(["Number", "Salary", "Name"], inplace = True)
nba.tail()
```

```python
# The previous sort_values method changed the position of the index
# To return to the original order, use the .sort_index method to sort by the index value
nba.sort_index()
```

## Rank value with the .rank() Method

```python
# Reset DataFrame
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.head(3)
```

```python
nba.sort_values(by = "Salary", ascending = False)
```
