```python
import pandas as pd
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# Alone, it will give an error but if you add an equal sign, it will create a new column
# if it does exist, it will replace the value
nba["Sport"] = "Basketball"  # scalar value. everything will have the value Basketball
```

```python
nba.head(3)
```

```python
nba["League"] = "National Basketball Association"
nba.head(3)
```

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# Broadcasting Operations
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# Broadcast to each 
nba["Age"].add(5)
nba["Age"] + 5  # shortcut of .add(5)
```

```python
nba["Salary"].sub(5000000)
nba["Salary"] - 5000000 # shortcut of .sub(5000000)
```

```python

nba["Weight"].mul(0.453592)  # multiply
nba["Weight in Kilograms"] = nba["Weight"] * 0.453592
```

```python
nba.head(3)
```

```python
nba["Salary"].div(1000000)
nba["Salary in Millions"] = nba["Salary"] / 1000000
```

```python
nba.head(3)
```
