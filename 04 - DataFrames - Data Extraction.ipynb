```python
import pandas as pd
```

```python
bond = pd.read_csv("jamesbond.csv")
bond.head(3)
```

## set_index() and reset_index() Methods

```python
# Instead of using index_col parameter during import, the index can be set later
bond.set_index(keys = "Film", inplace = True)
```

```python
bond.head(3)
```

```python
# To reset the DataFrame, use the reset_index() 
# parameter drop = False
# if drop = True, then the index will be dropped
bond.reset_index(drop = False, inplace = True)
bond.head(3)
```

```python
bond.set_index("Film", inplace = True)
bond.head(3)
```

```python
# Calling set_index again will drop and replace the index
bond.set_index("Year")
bond.head(3)
```

```python
# To prevent loss of column, use reset_index then set_index
bond.reset_index(inplace = True)
bond.set_index("Year", inplace = True)
```
