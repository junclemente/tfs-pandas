## The .agg() Method

```python
# Reset DataFrame
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby(["Sector", "Industry"])
fortune.head(3)
```

```python
# Using .agg() allows using different methods in each column
# Provide a dict:  key = column name, value = method
sectors.agg({"Revenue" : "sum",
            "Profits" : "sum", 
            "Employees" : "mean"})
```

```python
# Provide a list of methods to apply to each column
sectors.agg(["size", "sum", "mean"])
```

```python
# Another way to use .agg()
sectors.agg({"Revenue" : ["sum", "mean"],
            "Profits" : "sum", 
            "Employees" : "mean"})
```

## Iterating through Groups

```python
# Reset DataFrame
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby(["Sector", "Industry"])
fortune.head(3)
```

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for sector, data in sectors: 
    highest_revenue_comnpany_in_group = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_comnpany_in_group)
```

```python
df
```

```python
cities = fortune.groupby("Location")
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for city, data in cities:
    highest_revenue_in_city = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_in_city)
```
