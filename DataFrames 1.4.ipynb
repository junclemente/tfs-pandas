```python
import pandas as pd
```

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.dropna() # drops all rows that have a null value in any column
nba.dropna(how = "all", inplace = True) # removes all rows that have null value in all columns
nba.tail(3)
```

```python
nba.dropna(axis = 1)  # removes any column that has a null value in any position aka (axis = "columns")
```

```python
nba.dropna(subset = ["Salary", "College"])  # remove rows where null value is in specified column
```

```python
# fill in null values .fillna() Method
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.fillna(0)
```

```python
nba["Salary"].fillna(0, inplace = True)
nba.head()
```

```python
nba["College"].fillna("No College", inplace = True)

nba.head()
```

```python
# .astype() method
# requires all nulls
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
nba.head(6)
```

```python
nba.dtypes
nba.info()
```

```python
# pandas converts ints to floats when imported
# use .astype("int") to convert floats to integers
nba["Salary"] = nba["Salary"].astype("int")
nba.head()
```

```python
nba["Number"] = nba["Number"].astype("int")
nba["Age"] = nba["Age"].astype("int")  # can convert int to float => astype("float")
nba.info()
```

```python
nba["Position"].nunique()
```

```python
nba["Position"] = nba["Position"].astype("category")
nba.head(3)
nba.info()
```

```python
nba["Team"] = nba["Team"].astype("category")  # category datatype can help reduce size when a column has a small amount of unique values
nba.head()
```
