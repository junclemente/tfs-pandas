```python
import pandas as pd
```

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
# outer joins
week1.head(2)
```

```python
week2.head(2)
```

```python
# how = "outer"
# indicator: creates brand new column to show where value is pulled from 
merged = week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Wk 1", " - Wk 2"], indicator = True)
```

```python
merged["_merge"].value_counts()
```

```python
# use indicator to get values that are in either left or right but not both..
mask = merged["_merge"].isin(["left_only", "right_only"])
merged[mask]
```

```python
# left joins: includes all items in left DF and both left and right DFs. Does not include Right DF only... 
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.head(3)
```

```python
foods.head(3)
```

```python
week1 = week1.merge(foods, how = "left", on = "Food ID", sort = True)  # sorted by value in which matching ("Food ID")
```

```python
week1
```

```python
# left_on, right_on Parameters: when column names do not match... 
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week2.head(3)
```

```python
customers.head(3)
```

```python
# Drop will drop the ID column
week2 = week2.merge(customers, left_on = "Customer ID", right_on = "ID", sort = True).drop("ID", axis = 1) 
```

```python
week2.head()
```

```python
# merging by indexes: left_index, right_index parameters
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv", index_col = "ID")
foods = pd.read_csv("Restaurant - Foods.csv", index_col = "Food ID")
```

```python
foods.head(3)
```

```python
customers.head(3)
```

```python
week1.head()
```

```python
sales = week1.merge(customers, how = "left", left_on = "Customer ID", right_index = True)
```

```python
sales.head()
```

```python
foods.head()
```

```python
sales = sales.merge(foods, how = "left", left_on = "Food ID", right_index = True)
```
