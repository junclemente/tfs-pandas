```python
import pandas as pd
```

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.sort_values("Name", ascending = False)
nba.sort_values("Age", ascending = False)
nba.sort_values("Salary", ascending = False)
nba.sort_values("Salary", na_position = "first")  # na_position = "first" places null values first
```

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.sort_values(["Team", "Name"], ascending = False)
```

```python
nba.sort_values(["Team", "Name"], ascending = [True, False], inplace = True)  # list of boolean assigns sort there Team ascending, Name descending
nba.head()
```

```python
# sort_index() Method
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.sort_values(["Number", "Salary", "Name"], inplace = True)
nba.tail(3)
```

```python
nba.sort_index(ascending = False, inplace = True)
nba.tail()
```

```python
# Rank values with .rank() Method
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0). astype("int")
nba.head(3)
```

```python

nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.head(3)
```

```python
nba.sort_values(by = "Salary", ascending = False)
```
